Using the data in Excel file Loans, construct a scat-ter chart for monthly income versus loan amount and add a linear trendline. What is the regression model? If an individual has 7336 as monthly income, what would you predict the loan amount to be?



```python
import numpy as np
import math as mt
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (15, 8.0)
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
data=pd.read_excel("/home/dai/Downloads/Data_Files-20191003T101052Z-001/Data_Files/Loans.xlsx")
data
```

```python
X = data['Monthly_Income'].values
Y = data['Ln_Orig'].values

mean_x = np.mean(X)
mean_y = np.mean(Y)
n = len(X)
max_x = np.max(X) 
min_x = np.min(X)  
numer = 0
denom = 0
for i in range(n):
    numer += (X[i] - mean_x) * (Y[i] - mean_y)
    denom += (X[i] - mean_x) ** 2
slope = numer / denom
intercept = mean_y - (slope * mean_x)

x = np.linspace(min_x, max_x, 10)
y = slope * x + intercept
plt.plot(x, y, color= 'r', label='Regression Line')


X = X.reshape((n, 1))
reg = LinearRegression()
reg = reg.fit(X, Y)

    
plt.scatter(X, Y, color= 'g', label='Regression Line')
plt.xlabel('Monthly_Income')
plt.ylabel('Ln_Orig')


y = slope * 7336 + intercept
print(y)
```
